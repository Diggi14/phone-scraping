```python
import pandas as pd
import numpy as np
```

```python
phones=pd.read_csv('phones_unclean.csv')
phones.head(2)
phones.drop('Unnamed: 0',axis=1,inplace=True)
```

```python
phones['price']=phones['price'].str.replace('₹','').str.replace(',','').astype('Int32')
```

```python
phones['Spec-score']=phones['Spec-score'].astype('int16')
```

```python
phones=phones[phones['price']>3600] #removing feature phones
```

```python
phones.info()
```

```python
phones['Rating']=phones['Rating'].apply(lambda x:x.split(':')[1].split(';')[0]).astype('Float32')
```

```python
a=phones['model'].str.split('(')
```

```python
phones['model']=phones['model'].apply(lambda x:x.split('(')[0])
```

```python
phones[['brand', 'model']]=phones['model'].str.split(' ', n=1,expand=True)
```

```python
phones=phones[['brand','model',	'price','Spec-score','Rating','Network','processor','memory','battery','display','camera','memory_card','android_version']]
```

```python
phones.loc[indexes,['camera','memory_card']]=phones.loc[indexes,['camera','memory_card']].shift(periods=-1,axis=1).fillna(np.nan)
```

```python
phones[['display_size','other']]=phones['display'].str.split(',', n=1,expand=True)
```

```python
phones['display_size']=phones['display_size'].str.replace('inches','').str.replace('Display','').astype('Float32')
```

```python
phones.drop('display',axis=1,inplace=True)
```

```python
phones[['Resolution','Frame Rate']]=phones['other'].str.split(',',expand=True)
```

```python
phones[~phones['Resolution'].str.contains('px',na=False)]
```

```python
phones[['RAM','Memory']]=phones['memory'].str.split(',',expand=True)
```

```python
ind=phones[~phones['RAM'].str.contains('RAM',na=False)].index
phones.loc[ind,['RAM','Memory']]=phones.loc[ind,['RAM','Memory']].shift(axis=1).fillna("Unknown")
```

```python
phones[['Rear Camera','Front Camera']]=phones['camera'].str.split('&',expand=True)
```

```python
phones[~phones['Rear Camera'].str.contains('Rear',na=False)]
```

```python
phones.loc[881,['Front Camera']]=phones.loc[881,['Front Camera']].fillna('No camera')
```

```python
phones.drop(columns='camera',inplace=True)
```

```python
phones[['Battery','Fast charge']]=phones['battery'].str.split('with',expand=True)
```

```python
ind=phones[~phones['Fast charge'].str.contains('Fast',na=False)].index
phones.loc[ind,['Fast charge']]=phones.loc[ind,['Fast charge']].fillna("Does not support")
```

```python
phones.drop(columns='battery',inplace=True)
```

```python
phones.loc[502,['Battery']]='unknown'
phones['Battery']=phones['Battery'].apply(lambda x:x.split('mAh')[0])
```

```python
phones.loc[502,['Battery']]=np.nan
phones['Battery']=phones['Battery'].astype('float16')
```

```python
phones.to_csv('phone_clean.csv')
```
