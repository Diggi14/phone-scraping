```python
def bmi_category(bmi):
    if bmi < 18.5:
        return 'Underweight'
    elif bmi < 25:
        return 'Normal'
    elif bmi < 30:
        return 'Overweight'
    else:
        return 'Obese'

df['bmi_category'] = df['bmi'].apply(bmi_category)
```

```python
df.head(3)
```

```python
df['bmi_category'].value_counts().plot(kind='pie',autopct='%0.1f%%')
```

```python
sns.boxplot(data=df,x='bmi_category',y='age')
```

conclusion: median of all category are close not much distributed

```python
sns.violinplot(data=df,x='bmi_category',y='age')
```

```python
def bp_category(bp):
    if bp <= 120:
        return 'Normal'
    elif bp <= 129:
        return 'Elevated'
    elif bp < 140:
        return 'High BP 1'
    elif bp <= 180:
        return 'High BP 2'
    else:
        return 'Hypertensive'

df['bp_category'] = df['bloodpressure'].apply(bp_category)
```

```python
df.groupby('bp_category')['age'].agg(['mean','median'])
```
